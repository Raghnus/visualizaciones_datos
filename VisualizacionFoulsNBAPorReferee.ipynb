```python
df_play=df.loc[(df['Season type']=='Playoffs'),:]
df_regular=df.loc[(df['Season type']!='Playoffs'),:]
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,2);
plt.xlabel('Temporada',fontsize=16)
plt.ylabel('Fouls totales',fontsize=16)
plt.title('Fouls totales en Playoffs', fontsize=18)
plt.plot(df_play.groupby(['Season'])['Total fouls'].sum().index,
         df_play.groupby(['Season'])['Total fouls'].sum().values, 
         c='#5F0D0D',marker='*', markersize=15);
plt.subplot(1,2,1);
plt.xlabel('Temporada',fontsize=16)
plt.ylabel('Fouls totales',fontsize=16)
plt.title('Fouls totales en Fase Regular', fontsize=18)
plt.plot(df_regular.groupby(['Season'])['Total fouls'].sum().index,
         df_regular.groupby(['Season'])['Total fouls'].sum().values,
         c='#0F90AD',marker='*', markersize=15);
```
